# Análisis Exploratorio de Datos: Precios de Casas

Se realiza un análisis exploratorio del dataset **California Housing**.



Variable objetivo:

`MedHouseVal`: precio medio de la vivienda, expresado en cientos de miles de dólares.


## 1. Importación de bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
```

## 2. Carga del dataset

```python
dataset = fetch_california_housing(as_frame=True)

datos = dataset.frame

datos.head()
```

## 3. Descripción general del dataset

```python
print("Número de filas y columnas:")
print(datos.shape)

print("\nColumnas del dataset:")
print(datos.columns.tolist())
```

```python
datos.info()
```

## 4. Significado de las variables

| Variable | Descripción |
|---|---|
| MedInc | Ingreso medio de la zona |
| HouseAge | Edad media de las casas |
| AveRooms | Promedio de habitaciones por vivienda |
| AveBedrms | Promedio de recámaras por vivienda |
| Population | Población de la zona |
| AveOccup | Promedio de ocupantes por vivienda |
| Latitude | Latitud |
| Longitude | Longitud |
| MedHouseVal | Precio medio de la vivienda |


## 5. Revisión de valores nulos

```python
valores_nulos = datos.isnull().sum()

valores_nulos
```

## 6. Estadísticos descriptivos

```python
datos.describe().T
```

## 7. Distribución de la variable objetivo

La variable objetivo es `MedHouseVal`, que representa el precio medio de la vivienda.


```python
plt.figure(figsize=(8, 5))
plt.hist(datos["MedHouseVal"], bins=30)
plt.xlabel("Precio medio de vivienda")
plt.ylabel("Frecuencia")
plt.title("Distribución de la variable objetivo: MedHouseVal")
plt.tight_layout()
plt.show()
```

## 8. Distribución de las variables numéricas

```python
columnas_numericas = datos.columns

for columna in columnas_numericas:
    plt.figure(figsize=(8, 5))
    plt.hist(datos[columna], bins=30)
    plt.xlabel(columna)
    plt.ylabel("Frecuencia")
    plt.title(f"Distribución de {columna}")
    plt.tight_layout()
    plt.show()
```

## 9. Matriz de correlación

La correlación permite identificar qué variables tienen una relación lineal más fuerte con la variable objetivo.


```python
matriz_correlacion = datos.corr()

matriz_correlacion
```

```python
plt.figure(figsize=(10, 8))
plt.imshow(matriz_correlacion)
plt.colorbar()
plt.xticks(range(len(matriz_correlacion.columns)), matriz_correlacion.columns, rotation=90)
plt.yticks(range(len(matriz_correlacion.columns)), matriz_correlacion.columns)
plt.title("Matriz de correlación")
plt.tight_layout()
plt.show()
```

## 10. Correlación con la variable objetivo

```python
correlacion_objetivo = matriz_correlacion["MedHouseVal"].sort_values(ascending=False)

correlacion_objetivo
```

```python
correlacion_objetivo_sin_objetivo = correlacion_objetivo.drop("MedHouseVal")

plt.figure(figsize=(8, 5))
plt.barh(
    correlacion_objetivo_sin_objetivo.index,
    correlacion_objetivo_sin_objetivo.values
)
plt.xlabel("Correlación")
plt.ylabel("Variable")
plt.title("Correlación de cada variable con MedHouseVal")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## 11. Relación entre ingreso medio y precio de vivienda

Se espera que `MedInc` tenga una relación importante con `MedHouseVal`.


```python
plt.figure(figsize=(8, 5))
plt.scatter(datos["MedInc"], datos["MedHouseVal"], alpha=0.3)
plt.xlabel("Ingreso medio")
plt.ylabel("Precio medio de vivienda")
plt.title("Relación entre ingreso medio y precio de vivienda")
plt.tight_layout()
plt.show()
```

## 12. Relación geográfica: latitud, longitud y precio

Esta gráfica permite observar patrones espaciales en los precios de las viviendas.


```python
plt.figure(figsize=(8, 6))
plt.scatter(
    datos["Longitude"],
    datos["Latitude"],
    c=datos["MedHouseVal"],
    alpha=0.4
)
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.title("Distribución geográfica del precio medio de vivienda")
plt.colorbar(label="MedHouseVal")
plt.tight_layout()
plt.show()
```

## 13. Detección visual de valores extremos

Se utilizan diagramas de caja para observar posibles valores atípicos.


```python
for columna in datos.columns:
    plt.figure(figsize=(8, 4))
    plt.boxplot(datos[columna], vert=False)
    plt.xlabel(columna)
    plt.title(f"Diagrama de caja de {columna}")
    plt.tight_layout()
    plt.show()
```
