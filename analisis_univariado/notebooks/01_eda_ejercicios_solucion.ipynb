## 5. Medidas de tendencia central

Se tiene una variable **`fare`**, lo cuál indica cuánto pago un pasajero


### Actividad 5
Guarda la variable `fare` sin valores nulos en una nueva variable.

Usa `dropna()`.


```python
fare_sin_na = df["fare"].dropna()
print("Variable fare sin nulos (cantidad de valores):", len(fare_sin_na))
print(fare_sin_na.head(10))
```

### Actividad 6
Calcula:
- media de `fare`
- mediana de `fare`
- moda de `fare`

Usa `.mean()`, `.median()` y `.mode()`.


```python
media_fare = fare_sin_na.mean()
mediana_fare = fare_sin_na.median()
moda_fare = fare_sin_na.mode()

print("Media de fare:", media_fare)
print("Mediana de fare:", mediana_fare)
print("Moda de fare:")
print(moda_fare)
```

```python
media_fare = fare_sin_na.mean()
mediana_fare = fare_sin_na.median()
moda_fare = fare_sin_na.mode().iloc[0]

labels = ["Media", "Mediana", "Moda"]
values = [media_fare, mediana_fare, moda_fare]

plt.figure(figsize=(8, 5))
plt.bar(labels, values)
plt.title("Media, Mediana y moda")
plt.ylabel("Valor")

for i, value in enumerate(values):
    plt.text(i, value, f"{value:.2f}", ha="center", va="bottom")

plt.show()
```

### Actividad 7
Interpreta tus resultados.

1. ¿La media y la mediana son parecidas o diferentes?
2. ¿Qué podría indicar esa diferencia?
3. ¿Qué representa la moda en este caso?

**Respuesta:**  
1. La media es mayor que la mediana en este caso, lo que indica una distribución sesgada hacia valores altos de tarifa.
2. Esa diferencia sugiere que hay pasajeros que pagaron tarifas muy altas (valores atípicos a la derecha) que empujan la media hacia arriba, mientras que la mediana es más robusta frente a esos valores extremos.
3. La moda representa el valor de la tarifa que que más se repiten en el conjunto de datos, es decir, el valor de `fare` más común que fue de 8 libras.

## 6. Medidas de dispersión

**¿los valores están muy juntos o muy separados?**


### Actividad 8
Calcula para `fare`:
- mínimo
- máximo
- rango manual (`max - min`)
- rango con `np.ptp`
- varianza
- desviación estándar

Usar `.min()`, `.max()`, `.var()`, `.std()` y `np.ptp(...)`.


```python
min_fare = fare_sin_na.min()
max_fare = fare_sin_na.max()

rango_manual = max_fare - min_fare
rango_np = np.ptp(fare_sin_na)

varianza_fare = fare_sin_na.var()
desv_estandar_fare = fare_sin_na.std()

print("Mínimo:", min_fare)
print("Máximo:", max_fare)
print("Rango (manual):", rango_manual)
print("Rango (np.ptp):", rango_np)
print("Varianza:", varianza_fare)
print("Desviación estándar:", desv_estandar_fare)
```

### Actividad 9
Interpretación:

1. ¿La variable `fare` tiene mucha o poca variabilidad?
2. ¿Qué indica el rango?
3. ¿Qué interpretación le darías a la desviación estándar?



**Respuestas:**  
1. La variable `fare` muestra bastante variabilidad: hay tarifas muy bajas y otras muy altas, y la desviación estándar es grande en relación con la media.
2. El rango indica la diferencia entre la tarifa mínima y la máxima, es decir, el intervalo completo de valores observados para `fare`.
3. La desviación estándar mide, en promedio, cuánto se alejan las tarifas individuales del valor medio, un valor alto implica que muchas observaciones están lejos de la media y que la dispersión es grande.

## 9. Índice de Gini de heterogeneidad

Este índice se usa con variables categóricas para medir qué tan repartidas están las categorías.



### Actividad 16
Exploa la columna `class`:

1. Realiza el cálculo del índice Gini para la variable class




```python
proporciones_class = df["class"].value_counts(normalize=True)
gini_class = 1 - np.sum(proporciones_class**2)

print(proporciones_class)

k = len(df["class"].unique())

gini_max = 1 - 1/k
print("Gini max: " , gini_max)
print("ìndice de heterogeneidad Gini: ", gini_class)
```

### Actividad 17
Interpreta el resultado del índice de Gini.

- ¿El valor obtenido sugiere mucha o poca heterogeneidad y por qué?
- ¿Qué pasaría si una sola categoría concentrara casi todos los casos?


**Respuestas:**  
El índice de Gini calculado para `class` tiene un valor intermedio-alto: indica que existe heterogeneidad (hay varias clases presentes y ninguna concentra el 100% de los casos), pero no es máxima porque la clase `Third` domina.
Si una sola categoría concentrara casi todos los casos, la suma de las proporciones al cuadrado sería cercana a 1 y el índice de Gini se aproximaría a 0, reflejando muy poca diversidad entre categorías.
En cambio, valores más altos de Gini indican que las proporciones están más repartidas entre las distintas clases, es decir, mayor heterogeneidad.

## 10. Resumen

### Actividad 18
Responde con tus propias palabras:

- La media y la mediana sirven para...
- La desviación estándar sirve para...
- Las frecuencias sirven para...
- El índice de Gini de heterogeneidad sirve para...

**Respuesta (resumen):**
- **La media y la mediana sirven para** resumir el centro de una distribución: la media es el promedio aritmético de los datos y la mediana es el valor central cuando se ordenan.
- **La desviación estándar sirve para** medir cuánto se dispersan los datos alrededor de la media; cuanto mayor es, más separados están los valores del promedio.
- **Las frecuencias sirven para** saber cuántos casos y qué proporción del total hay en cada categoría o intervalo, lo que permite comparar grupos de forma sencilla.
- **El índice de Gini de heterogeneidad sirve para** cuantificar qué tan repartidos están los casos entre categorías: valores cercanos a 0 indican concentración en pocas categorías, mientras que valores más altos indican mayor diversidad u heterogeneidad.

Utiliza la variable  `survived` u otra que sea de tu interés para calcular:

```python
freq_abs_survived = df["survived"].value_counts()
freq_rel_survived = df["survived"].value_counts(normalize=True)

proporciones_survived = freq_rel_survived
gini_survived = 1 - np.sum(proporciones_survived**2)

print("Frecuencia absoluta (survived):")
print(freq_abs_survived)
print("\nFrecuencia relativa / proporciones (survived):")
print(freq_rel_survived)
k = len(df["survived"].unique())

gini_max = 1 - 1/k
print("Gini max: " , gini_max)
print("\nÍndice de Gini (survived):", gini_survived)
```

**Interpretación (variable survived):**

La tabla de frecuencias de 'survived' muestra cuántos pasajeros sobrevivieron (1) y cuántos no (0). Las proporciones indican que aproximadamente un 38% sobrevivió y un 62% no, lo que refleja que la mayoría de los pasajeros falleció en el hundimiento.

El índice de Gini para 'survived' tiene un valor intermedio: las dos categorías (sobrevivió / no sobrevivió) están relativamente equilibradas, aunque hay cierta concentración en la categoría 'no sobrevivió'. Un Gini cercano a 0 indicaría casi todos en una sola categoría; un valor más alto indicaría proporciones más parecidas. En este caso, el valor refleja que no hubo un dominio absoluto de una sola categoría.
