## 9. Índice de Gini de heterogeneidad

Este índice se usa con variables categóricas para medir qué tan repartidas están las categorías.



### Actividad 16
Exploa la columna `class`:

1. Realiza el cálculo del índice Gini para la variable class




```python
# proporciones de la variable class
p = df["class"].value_counts(normalize=True)

# calcular Gini
gini = 1 - (p**2).sum()

print(f"Gini: {gini}")
```

### Actividad 17
Interpreta el resultado del índice de Gini.

1. ¿El valor obtenido sugiere mucha o poca heterogeneidad y por qué?
2. ¿Qué pasaría si una sola categoría concentrara casi todos los casos?


**Respuestas:**  
1. El valor de Gini = 0.59 sugiere heterogeneidad moderada, ya que las observaciones no están concentradas completamente en una sola categoría, pero tampoco están distribuidas de forma totalmente uniforme entre todas.
2. Si una sola categoría concentrara casi todos los casos, el índice de Gini sería cercano a 0, lo que indicaría alta concentración y muy poca heterogeneidad en la variable.
